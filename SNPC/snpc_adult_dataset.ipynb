```python
from snpc import SNPC
import pandas as pd
import numpy as np
```

```python
a_d = pd.read_csv('new_adult.csv', index_col = [0])
income = a_d['income']
inc = pd.DataFrame(np.array(income), columns= ['labels'])
a_d.drop('income', axis= 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
x_train, x_val, y_train, y_val = train_test_split(np.array(a_d)[:6000], np.array(inc)[:6000], test_size=0.3, random_state=42)
```

```python
model = SNPC(2)
model.fit(x_train, y_train)
```

```python
model.evaluate(x_val, y_val)
```

```python
model.predict_all(x_val)
```

```python
model.predict_all(x_val, return_scores=True)
```

```python
#try for mean initialization
trained_model = SNPC(num_prototypes_per_class=1, initialization_type='random')
prototypes,proto_labels = trained_model.fit(x_train, y_train)
trained_model.evaluate(x_val, y_val)
```
